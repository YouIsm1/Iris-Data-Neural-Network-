```python
import numpy as np

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder

from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
```

```python
iris_data = load_iris()  # load the iris dataset
```

```python
print('Example data: \n' + str(iris_data.data[:5]))
print('Example labels: ' + str(iris_data.target[:5]))

x = iris_data.data
y_ = iris_data.target.reshape(-1, 1)  # Convert data to a single column
```

```python
# One Hot encode the class labels with sparse output
encoder = OneHotEncoder(sparse_output=True)
y = encoder.fit_transform(y_)

# Convert sparse tensor to dense tensor
y = y.toarray().squeeze()

print(f"on a y apres  one hot encoding : {y}")
```

```python
# Split the data for training and testing
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.20)
```

```python
# Build the model
model = Sequential()

model.add(Dense(10, input_shape=(4,), activation='relu', name='fc1'))
model.add(Dense(10, activation='relu', name='fc2'))
model.add(Dense(3, activation='softmax', name='output'))

# Adam optimizer with learning rate of 0.001
optimizer = Adam(lr=0.001)
model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])

print('Neural Network Model Summary: ' + str(model.summary()))
```

```python
# Train the model
model.fit(train_x, train_y, verbose=2, batch_size=5, epochs=500)
```

```python
# Test on unseen data
results = model.evaluate(test_x, test_y)

print('Final test set loss: {:4f}'.format(results[0]))
print('Final test set accuracy: {:4f}'.format(results[1]))
```

```python
# Prdictions sur les donnes de test
predictions = model.predict(test_x)

# Convertir les prdictions en classes
predicted_classes = np.argmax(predictions, axis=1)

# Comparer les prdictions avec les classes relles
print('Prdictions: ' + str(predicted_classes))
print('Classes relles: ' + str(test_y))
```
